```python
import numpy as np
import pandas as pd

from pathlib import Path
from pycaret.classification import setup, compare_models
```

```python
DATA_DIR = Path.cwd().resolve().parent / "data"

df_data = pd.read_parquet(DATA_DIR / "02_data_ataques_cardiacos_transformados.parquet")   
```

# Seleccion de columnas

```python
df_data.columns
```

```python
df_data.info()
```

```python
target_col = 'result'
```

```python
#Creacion del setup
models_setup = setup(
    data = df_data,
    target = target_col,
    session_id = 42,

    #Parametros para el pre-procesamiento
    numeric_features = ['age','heart_rate','systolic_bp','diastolic_bp','blood_sugar','ckmb','troponin'],
    categorical_features = ['gender'],

    #estrategias de imputacion
    numeric_imputation= 'mean',
    categorical_imputation = 'mode',
)
```

```python
df_data['result'].value_counts()
```

```python
best_model = compare_models(
    include = ['lr','rf','gbc','lightgbm','knn','xgboost'],
    sort = 'Accuracy',
    n_select = 3
)
```

```python
from pycaret.classification import create_model, tune_model, predict_model

print("Creando el modelo Random Forest Classifier...")
gbc_model = create_model('rf')

print("\n")
print("------")
print("\n")

print("Realizando fine-tuning del modelo....")
gbc_tuned = tune_model(gbc_model)

print("\n")
print("------")
print("\n")

print("Resumen del modelo")
print(gbc_tuned)

print("\n")
print("------")
print("\n")

print("Predicciones en el conjunto/set de test")
internal_predictions = predict_model(gbc_tuned)

print("\n")
print("------")
print("\n")


print(internal_predictions.head())
```
