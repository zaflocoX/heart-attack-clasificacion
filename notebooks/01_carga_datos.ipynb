# Carga de datos

```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv('../data/Medicaldataset.csv')
df.info()
```

```python
df.head()
```

```python
df.Result.unique()
```

```python
df.Result.value_counts()
```

```python
columnas_a_usar = [
    'Age',
    'Gender',
    'Heart rate',
    'Systolic blood pressure',
    'Diastolic blood pressure',
    'Blood sugar',
    'CK-MB',
    'Troponin',
    'Result'
]
```

```python
df = df[columnas_a_usar]
df.info()
```

# Ajuste de columnas

```python
df.columns = df.columns.str.replace(' ','_').str.lower()
df.info()
```

```python
df = df.rename(columns={'systolic_blood_pressure':'systolic_bp','diastolic_blood_pressure':'diastolic_bp'})
df.info()
```

```python
df = df.rename(columns={'ck-mb': 'ckmb'})
```

```python
df['result'] = df['result'].replace({'negative': 0 ,'positive': 1 }).astype('int64')
```

```python
df.info()
```

```python
df
```

# Creacion del archivo

```python
ruta_datos = Path.cwd().resolve().parent / 'data'
ruta_datos
```

```python
ruta_archivo = ruta_datos / "01_data_ataques_cardiacos.parquet"

ruta_datos.mkdir(parents=True, exist_ok=True)

df.to_parquet(ruta_archivo, index=False)
```
